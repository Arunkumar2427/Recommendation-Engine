```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from difflib import get_close_matches
from joblib import dump
```

```python
df = pd.read_csv("movies.csv")
df1 = df.copy()
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
df1.dropna(subset = "genres",inplace=True)
```

```python
vect = TfidfVectorizer()
feature_vectors = vect.fit_transform(df1["genres"])
```

```python
for i in enumerate(feature_vectors.toarray()):
    print(i)
```

```python
similarity  = cosine_similarity(feature_vectors)
```

```python
pd.DataFrame(similarity)
```

```python
titles= df["title"].tolist()
```

```python
x = input("Enter the movie name :")

m_name = get_close_matches(x,titles)

movie_index = df1[df1["title"]==m_name[0]].index.tolist()

similarity_score = list(enumerate(similarity[movie_index][0]))

a1 = pd.DataFrame(similarity_score)

len(similarity_score)

a1.sort_values(by =1,ascending=False,inplace=True)

a1_index = a1.index

a1_index
print(m_name)
df1.iloc[a1_index[0:20]]["title"]
```
